```python
# importing trandformers and datasets libraries
!pip install -q transformers datasets

# importing load_dataset to help load our model later
from datasets import load_dataset

# autoTokenizer to convert text data to tokens
from transformers import AutoTokenizer
```

```python
dataset = load_dataset("WutYee/HarryPotter_books_1to7")
```

```python
print(dataset)    # to see what the dataset contains
```

```python
# the dataset hold train, val, and test dsets, with only text

inspecting = dataset['train'] # lets get the keys in the train dict

print(inspecting)
```

```python
harry_potter_context = dataset['train']['text']  # set all the text in the train ds as context

# print(harry_potter_context)
```

```python
# we imported autotokenizer already so we just proceed

tokenizer = AutoTokenizer.from_pretrained("deepset/bert-base-cased-squad2")
```

```python
# now calling the tokenizer on only the first row of our dataset
"""
returns model inputs;
input_ids: the numbers representing the tokens in the text.
token_type_ids: indicates which sequence a token belongs to if there is more than one sequence.
attention_mask: indicates whether a token should be masked or not.

"""
tokenizer(harry_potter_context[0])
```

```python
# now we can tokenizing the whole dataset with a map() function and an example line from Harry Porter book to speed up the process

example = "usual amount of neck, which came in very useful as she spent so much of her"

def tokenization(example):
  return tokenizer(example["text"])

dataset = dataset.map(tokenization, batched = True)
```
