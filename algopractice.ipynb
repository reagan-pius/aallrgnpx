### Section 1: Implementation of search and sort algorithms

```python
import time

# Searching Algorithms

def linear_search(arr, target):
    for i, num in enumerate(arr):
        if num == target:
            return i
    return -1

def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Sorting Algorithms

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr


def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr


def selection_sort(arr):
    for i in range(len(arr)):
        min_index = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
```

### Section 2: Performance measurement

#### Sort algorithms

```python
# Function to measure execution time of the sort algorithms
def measure_performance(algorithm, sizes):
    times = []
    for size in sizes:
        arr = list(range(0, 100001, 10000))[:size]  # Limiting the array size
        start_time = time.time()
        algorithm(arr)
        end_time = time.time()
        times.append(end_time - start_time)
    return times

# Input sizes for the sorting algorithms
sort_input_sizes = list(range(100, 100001, 10000))

# Time taken for the sorting algorithms
bubble_sort_times = measure_performance(bubble_sort, sort_input_sizes)
insertion_sort_times = measure_performance(insertion_sort, sort_input_sizes)
selection_sort_times = measure_performance(selection_sort, sort_input_sizes)

# Plotting
```

#### Search algorithms

```python
# Function to measure the execution time of the search algorithms
def measure_search_performance(search_func, array, target):
    start_time = time.time()
    search_func(array, target)
    end_time = time.time()
    elapsed_time = end_time - start_time
    return elapsed_time

# Input array for searching algorithms
array = list(range(100, 100001, 1000))
target = array[-1] 

# searching algorithms performance
linear_search_times = []
binary_search_times = []

for size in array:
    array_subset = list(range(1, size + 1))     # Will use this to plot the x axis
    
    linear_search_time = measure_search_performance(linear_search, array_subset, target)
    binary_search_time = measure_search_performance(binary_search, array_subset, target)
    
    linear_search_times.append(linear_search_time)
    binary_search_times.append(binary_search_time)

# Plotting
```

#### Plots

```python
import time, matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

# Plotting sorting algorithm performance
plt.subplot(1, 2, 1)  # Creates the first subplot
plt.plot(sort_input_sizes, bubble_sort_times, label='Bubble Sort')
plt.plot(sort_input_sizes, insertion_sort_times, label='Insertion Sort')
plt.plot(sort_input_sizes, selection_sort_times, label='Selection Sort')
plt.xlabel('Input Size')
plt.ylabel('Time Taken (s)')
plt.title('Sorting Algorithm Performance')
plt.legend()

# Plotting searching algorithm performance
plt.subplot(1, 2, 2)  # Creates the second subplot
plt.plot(array, linear_search_times, label='Linear Search')
plt.plot(array, binary_search_times, label='Binary Search')
plt.xlabel('Input Size')
plt.ylabel('Time Taken (s)')
plt.title('Search Algorithm Performance')
plt.legend()

plt.tight_layout()
plt.show()
```

### Section 3: Analysis

Section 3: Analysis  

From the plotted graphs:

For searching algorithms, binary search shows significantly better performance compared to linear search, especially for larger input sizes. This is expected as binary search has a time complexity of O(log n) while linear search has a time complexity of O(n).

For sorting algorithms, insertion sort tends to perform better for smaller input sizes. As the input size increases, bubble sort and selection sort become increasingly slower. Among the three sorting algorithms, insertion sort generally performs the best, followed by selection sort, and then bubble sort. However, all three sorting algorithms have a time complexity of O(n^2), making them inefficient for large input sizes compared to more efficient algorithms like merge sort or quick sort.

Overall, if the input data is relatively small and simplicity is preferred, insertion sort may be a suitable choice. However, for larger datasets, more efficient algorithms with better time complexity, such as merge sort or quick sort, should be considered.
